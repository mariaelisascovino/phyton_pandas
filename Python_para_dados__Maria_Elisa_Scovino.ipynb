```python
###1. Ao visualizar a base você percebeu que seria melhor alterar o nome dealgumas colunas.
### Mude o ‘ID’ para ‘Identificador’
df_saude_do_sono_estilo_vida.rename(columns={'ID':'Identificador'}, inplace=True)
df_saude_do_sono_estilo_vida.head()
```

```python
###mude a coluna ‘Ocupação’ para ‘Profissão’
df_saude_do_sono_estilo_vida.rename(columns={'Ocupação':'Profissão'}, inplace=True)
df_saude_do_sono_estilo_vida.head()
```

```python
### a coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria IMC’.
df_saude_do_sono_estilo_vida.rename(columns={'Categoria BMI':'Categoria IMC'}, inplace=True)
df_saude_do_sono_estilo_vida.head()
```

```python
### corrija o nome da coluna que indica a pressão sanguínea
df_saude_do_sono_estilo_vida.rename(columns={'Pressão sanguíneaaaa':'Pressão Sanguínea'}, inplace=True)
df_saude_do_sono_estilo_vida.head()
```

```python
###2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]
df_saude_do_sono_estilo_vida.groupby('Profissão')['Duração do sono'].agg(['mean', 'median', pd.Series.mode])

```

```python
###3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?
# Filtrar o DataFrame para pessoas que atuam com engenharia de software
engenheiros_software = df_saude_do_sono_estilo_vida[df_saude_do_sono_estilo_vida['Profissão'] == 'Eng. de Software']
# Contar o número total de engenheiros de software
total_engenheiros_software = engenheiros_software.shape[0]
total_engenheiros_software ###retorna 4
```

```python
# Filtrar o DataFrame para engenheiros de software que são obesos
engenheiros_obesos = engenheiros_software[engenheiros_software['Categoria IMC'] == 'Obesidade']


# Contar o número de engenheiros de software obesos
total_engenheiros_obesos = engenheiros_obesos.shape[0]
total_engenheiros_obesos ###retorna 1

```

```python
# Calcular a porcentagem de obesos
porcentagem_obesos = (total_engenheiros_obesos / total_engenheiros_software) * 100
porcentagem_obesos ###retorna 25.0

print(f"A porcentagem de pessoas que atuam com engenharia de software e são obesas é {porcentagem_obesos:.2f}%")
```

```python
###4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)

# Filtrar o DataFrame para advogados e representantes de vendas
profissao = ['Advogado(a)', 'Representante de Vendas']
df_filtrado = df_saude_do_sono_estilo_vida[df_saude_do_sono_estilo_vida['Profissão'].isin(profissao)]

# Calcular a média de duração do sono para advogados e representantes de vendas separadamente
media_sono_advogado = df_filtrado[df_filtrado['Profissão'] == 'Advogado(a)']['Duração do sono'].mean()
media_sono_representante_vendas = df_filtrado[df_filtrado['Profissão'] == 'Representante de Vendas']['Duração do sono'].mean()

print(f"A média de duração do sono para Advogado(a) é {media_sono_advogado:.2f} horas.")
print(f"A média de duração do sono para Representante de Vendas é {media_sono_representante_vendas:.2f} horas.")
print("Ser representante de vendas faz você dormir menos")

```

```python
###5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (Use o método ‘isin’, considere a média)

# Filtrar o DataFrame para medicina e enfermagem
profissao2 = ['Médico(a)', 'Enfermeiro(a)']
df_filtrado = df_saude_do_sono_estilo_vida[df_saude_do_sono_estilo_vida['Profissão'].isin(profissao2)]

# Calcular a média de duração do sono para médicos e enfermeiros separadamente
media_sono_medico = df_filtrado[df_filtrado['Profissão'] == 'Médico(a)']['Duração do sono'].mean()
media_sono_enfermeiro = df_filtrado[df_filtrado['Profissão'] == 'Enfermeiro(a)']['Duração do sono'].mean()

print(f"A média de duração do sono para Médico(a) é {media_sono_medico:.2f} horas.")
print(f"A média de duração do sono para Enfermeiro(a) é {media_sono_enfermeiro:.2f} horas.")
print("O Médico(a) tem menos horas de sono")
```

```python
###7. Descubra qual a profissão menos frequente no conjunto. (Use value_counts)

# Contar a frequência de cada profissão
profissao_counts = df_saude_do_sono_estilo_vida['Profissão'].value_counts()
profissao_counts

```

```python
###8. Quem tem maior pressão sanguínea média, homens ou mulheres?(Considere a média)

# Dividir os valores da coluna "Pressão sanguíneaaaa" em duas colunas
df_saude_do_sono_estilo_vida[['Pressão Sistólica', 'Pressão Diastólica']] = df_saude_do_sono_estilo_vida['Pressão Sanguínea'].str.split('/', expand=True)

# Exibir as primeiras linhas do dataframe atualizado
print(df_saude_do_sono_estilo_vida.head())

```

```python
# Converter as novas colunas para float
df_saude_do_sono_estilo_vida['Pressão Sistólica'] = df_saude_do_sono_estilo_vida['Pressão Sistólica'].astype(float)
df_saude_do_sono_estilo_vida['Pressão Diastólica'] = df_saude_do_sono_estilo_vida['Pressão Diastólica'].astype(float)

# Calcular a pressão arterial média (PAM) por ID
df_saude_do_sono_estilo_vida['PAM'] = (df_saude_do_sono_estilo_vida['Pressão Sistólica'] + 2 * df_saude_do_sono_estilo_vida['Pressão Diastólica']) / 3
df_saude_do_sono_estilo_vida.head()
```

```python
# Calcular a média dos valores da coluna "PAM" para os valores "Homem" da coluna "Gênero"
media_pam_homem = df_saude_do_sono_estilo_vida[df_saude_do_sono_estilo_vida["Gênero"] == "Homem"]["PAM"].mean()
media_pam_homem
```

```python
# Calcular a média dos valores da coluna "PAM" para os valores "Mulher" da coluna "Gênero"
media_pam_mulher = df_saude_do_sono_estilo_vida[df_saude_do_sono_estilo_vida["Gênero"] == "Mulher"]["PAM"].mean()
media_pam_mulher
```

```python
###9. É predominante entre os participantes dormir 8 horas por dia(considere usar Moda como medida)?

# Calcular a moda da coluna "Duração do sono"
moda_duracao_sono = df_saude_do_sono_estilo_vida["Duração do sono"].mode()[0]

# Exibir a moda calculada
print(f"A moda da coluna 'Duração do sono' é: {moda_duracao_sono}")
print("Não é predominante os participantes dormirem 8 h por dia")
```

```python
###10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

# Calcular a média dos passos dados por pessoas com frequências cardíacas acima de 70
media_passos_acima_70 = df_saude_do_sono_estilo_vida[df_saude_do_sono_estilo_vida["Frequência cardíaca"] > 70]["Passos diários"].mean()
media_passos_acima_70

# Calcular a média dos passos dados por pessoas com frequência cardíaca menor ou igual a 70
media_passos_igual_ou_menor_70 = df_saude_do_sono_estilo_vida[df_saude_do_sono_estilo_vida["Frequência cardíaca"] <= 70]["Passos diários"].mean()
media_passos_igual_ou_menor_70

print(f"Média dos passos dados por pessoas com frequências cardíacas acima de 70: {media_passos_acima_70}")
print(f"Média dos passos dados por pessoas com frequência cardíaca menor ou igual a 70: {media_passos_igual_ou_menor_70}")
print("Pessoas com frequência cardíaca acima de 70 dão menos passos que as com menor ou igual a 70.")
```
